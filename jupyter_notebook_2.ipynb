```python
import pandas as pd
import numpy as np
```

```python
assets_df = pd.read_csv('all_assets_clean_clean.csv')
assets_df
```

```python
assets_df = assets_df.reset_index(drop=True)
assets_df.columns
```

```python
assets_df = assets_df.drop('Unnamed: 0', axis=1)
```

```python
assets_df['date'] = pd.to_datetime(assets_df['date'])
```

```python
# btc_df = assets_df[(assets_df['asset'] == 'BTC') & (assets_df['date'] < '2025-01-01')].copy()
assets_clean = assets_df[['date', 'asset', 'close', 'moon_phase']].copy()
assets_clean = assets_clean.sort_values(by=['asset', 'date']).reset_index(drop=True)

#Identify Buy and Sell signals
signals_new_moon = assets_clean[assets_clean['moon_phase'].isin(['Full Moon', 'New Moon'])].copy()
signals_new_moon = signals_new_moon[signals_new_moon['moon_phase'] != signals_new_moon['moon_phase'].shift()]
signals_new_moon['signal'] = signals_new_moon['moon_phase'].map({'Full Moon': 'buy', 'New Moon': 'sell'})
signals_new_moon


```

```python
new_moon_str =signals_new_moon[signals_new_moon['signal'].shift() == 'buy'].copy()
new_moon_str['buy_date'] = signals_new_moon['date'].shift()
new_moon_str['buy_price'] = signals_new_moon['close'].shift()
new_moon_str['sell_date'] = signals_new_moon['date']
new_moon_str['sell_price'] = signals_new_moon['close']
new_moon_str['return_pct'] = (new_moon_str['sell_price'] - new_moon_str['buy_price']) / new_moon_str['buy_price']
new_moon_str['cumulative'] = (1 + new_moon_str['return_pct']).cumprod()
new_moon_str = new_moon_str.dropna()
```

```python
new_moon_str
```

```python
#Identify Buy and Sell signals
signals_full_moon = assets_clean[assets_clean['moon_phase'].isin(['Full Moon', 'New Moon'])].copy()
signals_full_moon = signals_full_moon[signals_full_moon['moon_phase'] != signals_full_moon['moon_phase'].shift()]
signals_full_moon['signal'] = signals_full_moon['moon_phase'].map({'New Moon': 'buy', 'Full Moon': 'sell'})
signals_full_moon
```

```python
full_moon_str =signals_full_moon[signals_full_moon['signal'].shift() == 'buy'].copy()
full_moon_str['buy_date'] = signals_full_moon['date'].shift()
full_moon_str['buy_price'] = signals_full_moon['close'].shift()
full_moon_str['sell_date'] = signals_full_moon['date']
full_moon_str['sell_price'] = signals_full_moon['close']
full_moon_str['return_pct'] = (full_moon_str['sell_price'] - full_moon_str['buy_price']) / full_moon_str['buy_price']
full_moon_str['cumulative'] = (1 + full_moon_str['return_pct']).cumprod()
full_moon_str = full_moon_str.dropna()
```

```python
full_moon_str
```

```python
moon_strategies = pd.concat([new_moon_str, full_moon_str])
moon_strategies
```

```python
# Drop the first NaN per asset
assets_df = assets_df.sort_values(by=['asset', 'date'])
assets_df['daily_return'] = assets_df.groupby('asset')['close'].pct_change()
assets_df = assets_df.dropna(subset=['daily_return'])

assets_df['year'] = assets_df['date'].dt.year

yearly_returns = assets_df.groupby(['asset', 'year'])['daily_return'].apply(lambda x: (1 + x).prod() - 1).reset_index(name='yearly_return')

# Add a 'date' column with each year's Dec 31st
yearly_returns['date'] = pd.to_datetime(yearly_returns['year'].astype(str) + '-12-31')

# Rearrange columns for Tableau compatibility
compounded_returns = yearly_returns[['date', 'asset', 'yearly_return']]
```

```python
compounded_returns
```

```python
compounded_returns.to_csv('yearly_returns.csv')
```

```python
moon_strategies['year'] = moon_strategies['date'].dt.year

yearly_returns_by_strategy = (
    moon_strategies.groupby(['asset', 'moon_phase', 'year'])['return_pct']
    .apply(lambda x: (1 + x).prod() - 1)
    .reset_index(name='yearly_compounded_return')
)

yearly_returns_by_strategy['date'] = pd.to_datetime(yearly_returns_by_strategy['year'].astype(str) + '-12-31')
yearly_returns_by_strategy = yearly_returns_by_strategy[['asset', 'date','moon_phase', 'yearly_compounded_return' ]]
```

```python
yearly_returns_by_strategy
```

```python
yearly_returns_by_strategy.to_csv('yearly_returns_moon.csv')
```

```python
# btc_yearly_returns = compounded_returns[(compounded_returns['asset'] == 'BTC')]
compounded_returns = compounded_returns.rename(columns={'yearly_return': 'yearly_compounded_return'})

compounded_returns

# btc_df = assets_df[(assets_df['asset'] == 'BTC') & (assets_df['date'] < '2025-01-01')].copy()
```

```python
yearly_returns_asset_moon = pd.concat([yearly_returns_by_strategy, compounded_returns])
yearly_returns_asset_moon
```

```python
yearly_returns_asset_moon.to_csv('yearly_returns_asset_moon.csv')
```
